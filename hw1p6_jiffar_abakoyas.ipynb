Making fake data in 2 dimensions

Assume two classes A and B. 

Data from class A is normally distributed in 2D around the mean μA = (−1, −1) with standard deviation σA = 1/8. 

Data from class B is normally distributed in 2D around the mean μB =(1,1) with standard deviation σB = 1/8. 


```python
# Use the sklearn function “sklearn.datasets.make_blobs” so make these blobs. 
# In all these cases seed your random numbers.
# a. Assume there are 250 class A points, and 250 class B points. 
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from pandas import DataFrame
import numpy as np

# generate 2d classification dataset
#class A centered at (-1,-1), Red
#class B centered at (1,1), Blue
clusters, label = make_blobs(n_samples=500, n_features =2, centers=[[-1.0, -1.0],[1,1]], \
                             random_state = 21, cluster_std = 1/8)
```

```python
# Use matplot lib to plot these points. 
# These clusters should be very well separated and any classifier should do very well on them. 
# Please plot the blobs with class A in blue and class B points as in red. 

#Plot the two classes of blobs in red and blue
colors = ['r' if x == 0 else 'b' for x in label]
plt.scatter(x=clusters[:,0], y=clusters[:,1], c=colors)

```

```python
# Not that you have overplotting so you cannot see that your blobs are much denser in the middle then at the edges. 
# Decrease the size of the dots and also use a small transparency alpha << 1 
#until you can clearly see dots in the core as been denser. Always beware of overplotting.


# Decrease alpha and size for cluster points
plt.scatter(x=clusters[:,0], y=clusters[:,1], c=colors, alpha=0.5, s =5)
```

```python
# You may not always have access to sklearn. Reado your results in a. using only numpy.random.randn. 
# Without using sklearn you should be able to achive the same results, albiet with more lines of code.

#Save the means
mu_a = [-1,-1]
mu_b = [1,1]

#Using numpy creating 250 2D blobs
clusters_a = 1/8 * np.random.randn(250,2) + mu_a
clusters_b = 1/8 * np.random.randn(250,2) + mu_b

plt.scatter(x=clusters_a[:,0], y=clusters_a[:,1],c='r')
plt.scatter(x=clusters_b[:,0], y=clusters_b[:,1],c='b')
plt.title("Two blobs centered apart")
plt.show()
```

```python
plt.scatter(x=clusters_a[:,0], y=clusters_a[:,1],c='r', alpha=0.5, s =5)
plt.scatter(x=clusters_b[:,0], y=clusters_b[:,1],c='b', alpha=0.5, s =5)
plt.title('The two blobs centered apart, better alpha and size')
plt.show()
```

```python
# Going back to using sklearn for generating data sets we will now make the data sets less well separated. 
# Increase the standard deviation for A and B to σA = σB = 4. 
# Again plot the results and becareful with overplotting. 
# It should be clear here that any classifier, no matter how clever, will get very poor perfomance on this data.

#New clusters with higher standard deviation of 4
clusters_2, label_2 = make_blobs(n_samples=500, n_features =2, centers=[[-1.0, -1.0],[1,1]], \
                             random_state = 21, cluster_std = 4)


#Plot new clusters
#Plot the two classes of blobs in red and blue
colors_2 = ['r' if x == 0 else 'b' for x in label_2]
plt.scatter(x=clusters_2[:,0], y=clusters_2[:,1], c=colors_2, alpha=0.7, s= 40)
plt.title("Both Blobs with std=4, not separated")
```

```python
# For our last example we want to create data which is clearly separable but not linearly separaable. 
# Use sklearn.datasets.make_circles to make two circles in 2D each with 500 points. 
from sklearn.datasets import make_circles

# Unfortuately we don’t want a little scircle in the middle but a normally disturbed blob. 
circles, circle_label = make_circles(n_samples=500, noise=1/8, random_state=None, factor=0.8)

# Use fancy indexing to isolate data on the outer circle only (class B). 

# Then generate a blob in the center (class A) which will have mean μA = (0, 0) and standard deviation σA = 1/8. 
# Again show this data set in red and blue for the classes, showing that they are well separated.
colors_3 = ['r' if x == 0 else 'b' for x in circle_label]

plt.scatter(x=circles[:,0], y= circles[:,1], c=colors_3)
plt.title('Both Classes A and B')
```

```python
#Isolate and plot outer circle only
red = circle_label == 0
green = circle_label == 1
#Plot the outer circle
plt.scatter(circles[red, 0], circles[red, 1], c="r")
plt.title("Class B Circle")
```

```python
# Then generate a blob in the center (class A) which will have mean μA = (0, 0) and standard deviation σA = 1/8. 
mu_a = [0,0]
std = 1/8
clusters_a = std * np.random.randn(250,2) + mu_a
plt.scatter(x=clusters_a[:,0], y=clusters_a[:,1],c='b')
plt.title('The class A blob by itself')
```

```python
# Again show this data set in red and blue for the classes, showing that they are well separated.
plt.scatter(x=clusters_a[:,0], y=clusters_a[:,1],c='b')
plt.scatter(circles[red, 0], circles[red, 1], c="r")
plt.title('Both the blue Class A blob and the red Class B circle')
plt.show()
```
